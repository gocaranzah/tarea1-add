*El archivo tiene la misma estructura que los anteriores pero difiere en el separador de campos, dado que el separador es una expresion regular, se empleará la biblioteca 're'*

```python
import re
dolar_2020_regex = pd.DataFrame()

df_url ="../data/raw/dolar_2020_sii_AA.dat"
#df_url ='/content/drive/My Drive/Magister_UDD/Análisis_Datos/tarea1-add/data/raw/dolar_2020_sii_AA.dat'

patron='(XAX|XBX|XZX|XaX|XÃ¡X|XYX|XuX|XiX|XhX|XtX|XjX)'
#patron='[a-zA-Zá-ú]{3}'

with open (df_url,'r') as f:
    for linea in f:
        linea=linea.rstrip()
        lista=(re.sub(patron, '|', linea)).split('|')   
        dolar_2020_regex = dolar_2020_regex.append(pd.DataFrame([lista]), ignore_index=True)



dolar_2020_regex = dolar_2020_regex.iloc[:, 0:13]


#pd.set_option('display.max_columns', None)        
#dolar_2020_regex.head(100)
dolar_2020_regex.head()
```

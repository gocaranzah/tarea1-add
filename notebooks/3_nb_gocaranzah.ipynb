Luego, extiende la variable `dia_de_la_semana` para todos los valores. Finalmente, crea una variable con las combinaciones `[dia,mes]` que tienen al menos un `NaN` que sea en día de semana (es decir, excluyendo Sábado y Domingo). Luego, cuenta cuantos `NaN`(incluyendo Sábados y Domingos) hay para cada combinación`[dia/mes]`. ¿Qué puedes inferir de los días con mayor cantidad de `NaN`?  

```python
var_dia_mes = []

lu= 0
ma= 0
mi= 0
ju= 0
vi= 0
sa= 0
do= 0

j = 0
k = 0
for i, row in df_mg.iterrows():

    if np.isnan(row.value) :
        k += 1
            
        if row.dia_de_la_semana !=  'Sabado' and row.dia_de_la_semana !=  'Domingo':
            var_dia_mes.append( (row['dia'],row['mes']) )
            j += 1

            
        if row.dia_de_la_semana  == 'Lunes' :
            lu += 1
            
        if row.dia_de_la_semana  == 'Martes' :
            ma += 1
        
        if row.dia_de_la_semana  == 'Miercoles' :
            mi += 1
            
        if row.dia_de_la_semana  == 'Jueves' :
            ju += 1
            
        if row.dia_de_la_semana  == 'Viernes' :
            vi += 1
            
        if row.dia_de_la_semana  == 'Sabado' :
            sa += 1
            
        if row.dia_de_la_semana  == 'Domingo' :
            do += 1            
            
print("cantidad de NaN con Sábados y Domingos: ",k)
print("cantidad de NaN sin Sábados y Domingos: ",j)

print("Lunes    : ", lu)
print("Martes   : ", ma)
print("Miercoles: ", mi)
print("Jueves   : ", ju)
print("Viernes  : ", vi)
print("Sabado   : ", sa)
print("Dimingo  : ", do)


```

**Respuesta:** Los primeros dias de la semana y el fin de semana acumulan la mayor cantidad de datos Nan

Calcula el promedio del precio del dolar en los días de cumpleaños del grupo para cada año entre 1983 y 2021 y el promedio del precio del dolar para cada año completo entre 1983 y 2021, y muestra una tabla donde se vean ambos resultados. ¿Qué diferencias hay en ambos indicadores en cada año?




```python
df_bc_all.head()
```

```python

df2 = pd.DataFrame()

for i  in range(1983,2022):
    p1 = df_bc_all[ (df_bc_all['year']==i ) &  ( (df_bc_all['mes'] == 2 ) &  (df_bc_all['dia'] == 6 ) |   (df_bc_all['mes'] == 7 ) &  (df_bc_all['dia'] == 1 )  ) ]['1.Dólar observado'].mean()
    p2 = df_bc_all[ (df_bc_all['year']==i )  ]['1.Dólar observado'].mean()

    nueva_fila = { 'año':i , 'grupo': p1, 'completo': p2} 
    df2 = df2.append(nueva_fila, ignore_index=True)

print (df2)

    
```

```python
import matplotlib.pylab as plt
df2.completo.plot.hist( figsize=(10,5),alpha=0.9)
df2.grupo.plot.hist( figsize=(10,5), alpha=0.8)
plt.legend()
```
